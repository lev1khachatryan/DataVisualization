### Bar Plot

```python
np.random.seed(42)
x_mean = np.random.randint(20, 40, 5)
x_std = np.random.randint(2, 5, 5)
```

```python
fig = plt.figure(figsize=(16,10))  
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel("Group")
ax.set_ylabel("Some Value")
ind = np.arange(5) # The position of the bars on the x-axis
ax.bar(ind, x_mean, yerr = x_std);
```

### Stacked Bar Plot

```python
np.random.seed(42)
x_mean = np.random.randint(20, 40, 5)
x_std = np.random.randint(2, 5, 5)

y_mean = np.random.randint(15, 35, 5)
y_std = np.random.randint(3, 6, 5)
```

```python
fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel("Group")
ax.set_ylabel("Some Value")

names = ["A", "B", "C", "D", "E"]
ind = np.arange(5)
plt.xticks(ind, names, fontweight='bold')

ax.bar(ind, x_mean, yerr = x_std, edgecolor='white');
ax.bar(ind, y_mean, bottom = x_mean, yerr = y_std, edgecolor='white');
```

### Percentage Stacked Bar Plot

```python
np.random.seed(42)
x_mean = np.random.randint(20, 40, 5)
x_std = np.random.randint(2, 5, 5)

y_mean = np.random.randint(15, 35, 5)
y_std = np.random.randint(3, 6, 5)

z_mean = np.random.randint(5, 15, 5)
z_std = np.random.randint(1, 2, 5)
```

```python
total = [i+j+k for i,j,k in zip(x_mean, y_mean, z_mean)]
```

```python
print(total)
```

```python
x_perc = [i / j * 100 for  i,j in zip(x_mean, total)]
y_perc = [i / j * 100 for  i,j in zip(y_mean, total)]
z_perc = [i / j * 100 for  i,j in zip(z_mean, total)]
```

```python
print(x_mean)
print(x_perc)
```

```python
fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(1, 1, 1)

ax.set_xlabel("Group")
ax.set_ylabel("Some Value")

names = ["A", "B", "C", "D", "E"]
ind = np.arange(5)
plt.xticks(ind, names, fontweight='bold')

ax.bar(ind, x_perc, yerr = x_std, edgecolor='white')
ax.bar(ind, y_perc, bottom = x_perc, yerr = x_std, edgecolor='white')
ax.bar(ind, z_perc, bottom = [i+j for i,j in zip(x_perc, y_perc)], yerr = y_std, edgecolor='white');
```

### Line Plot

```python
np.random.seed(42)
x = np.random.normal(1, 1, 100) + np.linspace(1, 15, 100)
```

```python
fig = plt.figure(figsize=(16,10))  
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel("Time")
ax.set_ylabel("Some Value")
ax.plot(x);
```

### Area Plot

```python
fig = plt.figure(figsize=(16,10))  
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel("Time")
ax.set_ylabel("Some Value")
ax.fill_between(np.linspace(1, 15, 100), x);
```

### Errorbar

```python
np.random.seed(42)
x = np.arange(1, 20, 1)
y = x**1.5
sd = np.random.normal(0, 5, 19)
```

```python
fig = plt.figure(figsize=(16,10))  
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Time')
ax.errorbar(x, y, yerr=sd, fmt = "--o");
```

### Multiple subplots

```python
np.random.seed(42)
x1 = np.random.exponential(10, 92) + np.linspace(1, 20, 92)
x2 = np.random.uniform(1, 25, 92) - np.linspace(1, 30, 92)
x3 = np.random.normal(1, 1, 92) - np.linspace(1, 15, 92)
x4 = np.random.normal(1, 2, 92)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 10));

# Set plot line width
line_width = 0.75

ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]

# Plot data
ax11.plot(x1, c='blue', lw=line_width)
ax12.plot(x2, c='orange', lw=line_width)
ax21.plot(x3, c='green', lw=line_width)
ax22.plot(x4, c='brown', lw=line_width)

ax21.set_xlabel('Time')
ax22.set_xlabel('Time')
ax11.set_ylabel('Sales')
ax21.set_ylabel('Sales')

ax11.text(85, np.max(x1)*0.99, 'Winter')
ax12.text(85, np.max(x2)*0.99, 'Spring')
ax21.text(85, np.max(x3)*0.99, 'Summer')
ax22.text(85, np.max(x4)*0.99, 'Fall');
```

### Japanese Candlesticks

```python
start_date = '2012-01-01'
end_date = '2012-03-01'
```

```python
df = data.DataReader('AAPL', 'yahoo', start_date, end_date)
```

```python
print(df.shape)
```

```python
df.head()
```

```python
op = df.Open
cl = df.Close
lo = df.Low
hi = df.High
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 10));

# Set plot line width
line_width = 0.75

ax11 = axes[0][0]
ax12 = axes[0][1]
ax21 = axes[1][0]
ax22 = axes[1][1]

# Plot data
ax11.plot(op, c='blue', lw=line_width)
ax12.plot(cl, c='orange', lw=line_width)
ax21.plot(lo, c='green', lw=line_width)
ax22.plot(hi, c='brown', lw=line_width)

ax21.set_xlabel('Time')
ax22.set_xlabel('Time')
ax11.set_ylabel('Price')
ax21.set_ylabel('Price');
```

```python
fig, ax = plt.subplots(figsize = (16, 10))
candlestick_ohlc(ax, 
                 zip(mdates.date2num(df.index.to_pydatetime()), op, hi, lo, cl),
                width = 0.5, colorup = "green")

ax.xaxis_date();
```
